```python
np.random.seed(0)

# Number of rows and columns
num_rows = 40
num_cols = 5

# Generating random data
data = {
    'A': np.random.normal(loc=50, scale=10, size=num_rows),  # Normal distribution
    'B': np.random.uniform(low=0, high=100, size=num_rows),  # Uniform distribution
    'C': np.random.gamma(shape=2, scale=10, size=num_rows),  # Gamma distribution
    'D': np.random.exponential(scale=10, size=num_rows),  # Exponential distribution
    'E': np.random.poisson(lam=5, size=num_rows),  # Poisson distribution
}

# Introducing null values
for col in data:
    null_indices = np.random.choice(num_rows, size=5, replace=False)
    data[col][null_indices] = 0

df = pd.DataFrame(data)
```

```python
data = {
    'name': ['abc', 'def', 'ghi', 'jkl', 'mno', 'pqr', 'stu', 'vwx', 'yz', 'ab', 'cd', 'ef', 'gh', 'ij'], 
    'age': [20, 12, np.nan, np.nan, 20, 11, 9, 1, 90, 44, 32, np.nan, 43, 44]
}
```

```python
age_mean = df['age'].mean()

df.fillna(age_mean, inplace=True)
df['age']
```

```python
outliers = []
import seaborn as sns

def detect_outliers(data, threshold):
    for i in df['age']:
        mean = df['age'].mean()
        std = df['age'].std()
        z_score = (i - mean) / std
        if np.abs(z_score) > threshold:
            print('Value beyond threshold : ',i,' with z_score : ',z_score)
            outliers.append(i)
    return outliers
print(detect_outliers(df,1))
```

```python

def outlier_limits(col):
    q3 = df['age'].quantile(0.75)
    q1 = df['age'].quantile(0.25)
    
    iqr = q3 - q1
    print(iqr," is the IQR")
    
    ul = q3 + 1.5 * iqr
    ll = q1 - 1.5 * iqr
    return ul, ll

for ele in df:
    ul, ll = outlier_limits(df[ele])
    print('Element : ',ele," ll : ",ll,' ul : ',ul)
    df[ele] = np.where( df[ele] > ul, ul, np.where( df[ele] < ll, ll, df[ele] ))
```
