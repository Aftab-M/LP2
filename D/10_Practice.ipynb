```python
iqr_outliers = []

def get_outliers_by_iqr(col):

    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    for i in df[col]:
        if i<lower_bound or i<upper_bound:
            print(f'Found the outlier {i}')
            iqr_outliers.append(i)

    return iqr_outliers

get_outliers_by_iqr('sepal_width')


```

```python
df['sepal_width'] = df['sepal_width']<3
df['sepal_width'] = df['sepal_width']>1
sns.boxplot(df)
```
