```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sna.csv')
```

```python
df['Genderr'] = df['Gender'].replace({'Male':0, 'Female':1})
df['Genderr']
```

```python
df.isna().sum()
```

```python
df['Purchased']
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score, recall_score, f1_score

X = df[['Genderr','Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=44)
```

```python
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
ypred
```

```python
ypred[:5], ytest.head()
```

```python
conf = confusion_matrix(ytest, ypred)
TN, FP, FN, TP = conf.ravel()

TN, FP, FN, TP
```

```python
acc = accuracy_score(ytest, ypred)
erate = 1 - acc
prc = precision_score(ytest, ypred)
recall = recall_score(ytest, ypred)
f1 = f1_score(ytest, ypred)

print('Error Rate : ', erate)
print('Precision : ', prc)
print('Accuracy : ', acc)
print('Recall : ', recall)
print('F1 Score : ', f1)
```
