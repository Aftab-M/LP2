```python
outliers_from_zscore = []



def get_outliers_by_Z(col, threshold):
    
    col_mean = df[col].mean()
    col_std = df[col].std()

    print('Mean for column is ',col_mean, ' and standard deviation is ', col_std)

    for i in df[col]:
        z_score = (i - col_mean) / col_std
        # print(z_score)
        if(abs(z_score) > threshold):
            print('Found outlier ',i, ' with z_score as ', z_score)
            outliers_from_zscore.append(i)

    return outliers_from_zscore






get_outliers_by_Z('Science_Score', 1)



```

```python

outliers_by_iqr = []


def get_outliers_by_iqr(col):

    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)

    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    print('Lower bound is ',lower_bound,' and upper bound is ',upper_bound)

    for i in df[col] : 
        if i < lower_bound or i > upper_bound:
            print('Found outlier : ',i)
            outliers_by_iqr.append(i)
            
get_outliers_by_iqr('Science_Score')





```
